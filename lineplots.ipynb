```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
loop_size = pd.read_csv("line_plot.csv")
loop_size.head()
```

```python
chr1_size = loop_size.query("Chromosome == 'chr1'")
chr6_size = loop_size.query("Chromosome == 'chr6'")
colors = ['#8b4513','#008000','#4682b4','#4b0082','#ff0000','#00ff00','#00ffff','#0000ff','#da70d6','#ff1493','#ff8c00']
palette = sns.color_palette(colors)
```

```python
g = sns.FacetGrid(chr6_size, col="Chromosome", hue="Tool", palette=palette, height=5)
g.map(sns.lineplot, "Resolution", "Size (kb)", size=5)
g.add_legend()
g.savefig("chr6_size_in_kb.png")
```

```python
g = sns.FacetGrid(chr1_size, col="Chromosome", hue="Tool", palette=palette, height=5)
g.map(sns.lineplot, "Resolution", "Size (kb)", size=2)
g.add_legend()
g.savefig("chr1_size_in_kb.png")
```

# Regression Plot

```python
regression = pd.read_csv("regression_pl.csv")
regression.head()
```

```python
chr1_regression = regression.query("Chromosome == 'chr1'")
chr6_regression = regression.query("Chromosome == 'chr6'")
colors = ['#8b4513','#008000','#4682b4','#4b0082','#ff0000','#00ff00','#00ffff','#0000ff','#da70d6','#ff1493','#ff8c00']
palette = sns.color_palette(colors)
```

```python
sns.lmplot(data=chr6_regression, x="Size (# bins) vs. Resolution", y="Size (kb) vs. Resolution", hue="Category", ci=None, palette=palette, aspect=2,
           scatter_kws={"s": 50, "alpha": 1});
plt.savefig("chr6_regression_categorical.png")
```

```python
sns.lmplot(data=chr1_regression, x="Size (# bins) vs. Resolution", y="Size (kb) vs. Resolution", hue="Category", ci=None, palette=palette, aspect=2,
           scatter_kws={"s": 50, "alpha": 1});
plt.savefig("chr1_regression_categorical.png")
```

```python
def label_point(x, y, val, ax):
    a = pd.concat({'x': x, 'y': y, 'val': val}, axis=1)
    for i, point in a.iterrows():
        ax.text(point['x']+0.5, point['y'], str(point['val']))
```

```python
ax = sns.lmplot(x="Size (# bins) vs. Resolution",
                y="Size (kb) vs. Resolution",
                data=chr1_regression,
                fit_reg=True,
                height=5,
                aspect=2) # size and dimension
plt.xlabel('Size (# bins) vs. Resolution')
plt.ylabel('Size (kb) vs. Resolution')
#label_point(chr1_regression["Size (# bins) vs. Resolution"], chr1_regression["Size (kb) vs. Resolution"], chr1_regression["Tool"], plt.gca())
#plt.figure(figsize=(10,20))
plt.savefig("chr1_regression.png")
```

```python
sns.lmplot(x="Size (# bins) vs. Resolution",
           y="Size (kb) vs. Resolution",
           data=chr6_regression,
           fit_reg=True,
           height=5,
           aspect=2)
plt.xlabel('Size (# bins) vs. Resolution')
plt.ylabel('Size (kb) vs. Resolution')
#label_point(chr6_regression["Size (# bins) vs. Resolution"], chr6_regression["Size (kb) vs. Resolution"], chr6_regression["Tool"], plt.gca())
plt.savefig("chr6_regression.png")
```
